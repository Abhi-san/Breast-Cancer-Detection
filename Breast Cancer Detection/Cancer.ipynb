```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
print(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize=(20,20))
sns.heatmap(data_frame.corr(), annot = True, cmap ='cividis', linewidths=2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras

tf.random.set_seed(3)

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")

def predict_tumor(input_data, model, scaler):
    # Reshape and standardize input data
    input_data_as_numpy_array = np.asarray(input_data)
    input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
    input_data_std = scaler.transform(input_data_reshaped)
    
    prediction = model.predict(input_data_std)
    prediction_label = np.argmax(prediction)

    return prediction_label
```

```python
input_data = [11.76, 21.6, 74.72, 427.9, 0.08637, 0.04966, 0.01657, 0.01115, 0.1495, 0.05888, 0.4062, 1.21, 2.635, 28.47, 0.005857, 0.009758, 0.01168, 0.007445, 0.02406, 0.001769, 12.98, 25.72, 82.98, 516.5, 0.1085, 0.08615, 0.05523, 0.03715, 0.2433, 0.06563]

# Call the predict_tumor function
prediction_label = predict_tumor(input_data, model, scaler)

if prediction_label == 0:
    print('The tumor is Malignant')
else:
    print('The tumor is Benign')
```
